This notebook demonstrates inference with `swyft` for $\Lambda$CDM cosmology with a CMB power spectra simulator. We:
- define the simulator, which based on `CLASS`
- load or define a data store
- define a linear compression network
- load or train inference
- plot weighted samples and compare to MCMC
- perform an empirical credibility test.

The defined simulator also computes some BAO values, which can be incorporated into the inference if desired. By default they are ignored, but this can be modified by changing the `Head` network. At the end of the notebook the simulator for varying the neutrino mass is defined.

Define the simulator and noise model! The true noise model is a correlated $\chi^2$ (i.e. Wishart) distribution with $f_{sky}(2\ell+1)$ degrees of freedom. Since simulating this is expensive at large $\ell$, we use a Gaussian approximation starting at $\ell=51$. Note also that to call the noise model, it is efficient to sample the Gaussian noise ahead of time and store it in memory.

```python
lmax=2500
fsky=0.6
ell=np.array([l for l in range(2,lmax+1)])
ells=ell*(ell+1)/(2*np.pi)

#detector contribution to signal, from file
Nltt=fpr2[ell-2,1]
Nlee=fpr2[ell-2,2]

sigBAO=np.array([0.015,0.17,0.13,0.16])
bao0={'6DF': [0.106, 0.327,0.015, 7], 'LOWZ': [0.32,8.47, 0.17,3],
            'CMASS':   [0.57,13.77,0.13,3], 'MGS':[0.15,4.47,0.16,3]}
def simulator(v,sigma=0.1):
    #to avoid some CosmoComputationErrors
    V=np.around(v, decimals=6)
    omega_b, omega_cdm, theta, lnAs,n_s,tau_reio=V[0],V[1],V[2],V[3],V[4],V[5]
    A_s=np.exp(lnAs)/ 1.e10
    params = {
        'output': 'tCl pCl lCl',
        'l_max_scalars': lmax,
        'lensing': 'yes',
        'omega_b': omega_b,
        'omega_cdm': omega_cdm,
        '100*theta_s': theta,
        'A_s': A_s,
        'n_s': n_s,
        'tau_reio': tau_reio
    }

    # Create an instance of the CLASS wrapper
    cosmo = Class()

    # Set the parameters to the cosmological code
    cosmo.set(params)

    # Run the code.
    
    try:
        cosmo.compute(['lensing'])
        # Extract observables.
        cls = cosmo.lensed_cl(lmax)
        T=cosmo.T_cmb()
        baoT=[]
        for i,k in enumerate(bao0.keys()):
            z=bao0[k][0]
            da = cosmo.angular_distance(z)
            dr = z / cosmo.Hubble(z)
            dv = pow(da * da * (1 + z) * (1 + z) * dr, 1. / 3.)
            rs = cosmo.rs_drag()

            if bao0[k][-1] == 3:
                theo = dv / rs
            elif bao0[k][-1] == 4:
                theo = dv
            elif bao0[k][-1] == 5:
                theo = da / rs
            elif bao0[k][-1] == 6:
                theo = 1. / cosmo.Hubble(z) / rs
            elif bao0[k][-1] == 7:
                theo = rs / dv
            baoT.append(theo)
        cosmo.struct_cleanup()
        return dict(TT=cls['tt'][ell]*(T*1.e6)**2+Nltt, TE=cls['te'][ell]*(T*1.e6)**2,\
                EE=cls['ee'][ell]*(T*1.e6)**2+Nlee, BAO=np.array(baoT))
    except:
        print("a simulation failed! Find a 3 in store.get_simulation_status")
        return dict(TT=np.nan, TE=np.nan, EE=np.nan, BAO=np.nan)


#noise from cosmic variance
#before calling this function, need to generate noisevecs

def noise(obs, params=None, sigma=1.0):
    cat=np.concatenate((obs['TT'],obs['EE'],obs['TE']),axis=0)
    CLTT=[]
    CLEE=[]
    CLTE=[]
    #need to modify degrees at low ell of freedom, including fsky
    for l in range(2,2+49):
        #below is modified d.o.f. version
        nm=int(np.round(fsky*(2*l+1)))
        g1=np.random.randn(nm)
        g2=np.random.randn(nm)
        almT=np.sqrt(obs['TT'][l-2])*g1
        almE=(obs['TE'][l-2]/np.sqrt(obs['TT'][l-2]))*g1+np.sqrt(obs['EE'][l-2]-obs['TE'][l-2]**2 / obs['TT'][l-2])*g2

        CLTT.append((((nm))**-1 *np.sum((almT**2))))#*np.array([1]+[2]*l)))))
        CLEE.append((((nm))**-1 *np.sum((almE**2))))#*np.array([1]+[2]*l)))))
        CLTE.append((((nm))**-1 *np.sum((almT*almE))))#*np.array([1]+[2]*l)))))
        
    NN=noisevecs[np.random.randint(0,100000)]
    noise=np.concatenate((np.zeros(29+20), NN[29+20:2499], np.zeros(29+20), NN[2528+20:4998], np.zeros(29+20),
                         NN[5027+20:]))
    cat[:49]=np.array(CLTT)
    cat[2499:2499+49]=np.array(CLEE)
    cat[4998:4998+49]=np.array(CLTE)
    cat=cat+noise
    
    return {'TT': cat[:2499],'EE': cat[2499:4998], 'TE': cat[4998:], 
            'BAO': obs['BAO']+np.random.randn(*obs['BAO'].shape)*sigBAO}
```

now generate reasonable prior (just CMB) using Fisher analysis. Start by defining finite derivative

```python
def finiteD(k, eps=1e-2):
    parr=list(pl)
    parr[k]=pl[k]*(1+eps)
    parr2=list(pl)
    parr2[k]=pl[k]*(1-eps)
    m1=simulator(parr)
    m2=simulator(parr2)
    output=(m1['TT']-m2['TT'])/(pl[k]*(2*eps))
    for kk in ['EE','TE']:#,'BAO']:
        output=np.concatenate((output,(m1[kk]-m2[kk])/(pl[k]*(2*eps))),axis=0)
    return output
```

compute derivatives and Fisher matrix

```python
ders=[finiteD(i) for i in range(6)]
```

```python
F=np.zeros((6,6))
for i in range(6):
    for j in range(6):
        d1=ders[i]
        if i != j: 
            d2=ders[j]
        else: 
            d2=d1
        term1=np.dot(d1,np.matmul(icovmat,d2))
        #term2=np.sum(d1[-4:]*d2[-4:]/((sigBAO)**2))
        F[i,j]=term1#+term2
```

```python
Finv=np.linalg.inv(F)
```

we can check our approximation

```python
#this is "exact" version of log-likelihood
def chi2(fid, test):
    detTestL=test['TT']*test['EE']-test['TE']**2
    detFidL=fid['TT']*fid['EE']-fid['TE']**2
    #print(test['TT'][0], test['EE'][0], test['TE'][0])
    #print(fid['TT'][0], fid['EE'][0], fid['TE'][0])
    #print(detTestL[0])
    #print(detFidL[0])
    mix=fid['TT']*test['EE']+test['TT']*fid['EE']-2*fid['TE']*test['TE']
    res=np.sum(fsky*(2*ell+1)*(mix/detTestL+np.log(detTestL/detFidL)-2))
    return res
def lnL(fid,test):
    return -0.5*chi2(fid,test)
```

```python
#use this as consistency check for fisher 
def finiteD2LNL(k, eps=1e-2):
    parr=list(pl)
    parr[k]=pl[k]*(1+eps)
    parr2=list(pl)
    parr2[k]=pl[k]*(1-eps)
    m1=simulator(parr)
    m2=simulator(parr2)
    m0=simulator(pl)
    return (lnL(m0,m1)+lnL(m0,m2)-2*lnL(m0,m0))/(pl[k]*eps)**2
```

now define linear compression networks.

```python
# if noise is not included in the observation,
# this compression network gives accurate central values
# for the parameters. This subtlety arises because the
# likelihood at low-\ell is non-Gaussian.
from swyft.networks import OnlineNormalizationLayer
import torch.nn.functional as F
class Head(swyft.Module):
    def __init__(self, obs_shapes):
        super().__init__(obs_shapes=obs_shapes)

        self.n_features = 15
        self.l = torch.nn.Linear(7497+4, 10)
        self.l2 = torch.nn.Linear(98,5)
        self.onl_f = OnlineNormalizationLayer(torch.Size([7497+4]))
        self.onl_f2 = OnlineNormalizationLayer(torch.Size([98]))
        #self.Ells = torch.tensor(ells, dtype=torch.float32).to('cuda')

    def forward(self, obs):
        #print(obs)
        x = torch.cat((obs['TT'][:,49:], obs['EE'][:,49:], obs['TE'], obs['BAO']), axis=1)
        #x = torch.cat((obs['TT'][:,49:], obs['EE'][:,49:], obs['TE']), axis=1)
        x1 = torch.sqrt(torch.cat((obs['TT'][:,:49],
                        obs['EE'][:,:49]), axis=1))
        x=torch.cat((x,x1),axis=1)
        x1=torch.sqrt(x1)
        x=self.onl_f(x)
        x1=self.onl_f2(x1)
        nbatch = len(obs)
        x = self.l(x)
        x1=self.l2(x1)

        return torch.cat((x,x1),axis=1)
        #return x
```

```python
# if noise is not included in the observation,
# this compression network gives accurate central values
# for the parameters. This subtlety arises because the
# likelihood at low-\ell is non-Gaussian.
# if you really want to compare against MCMC forecasting,
# either (i) use an observation with noise or (ii) use this network.
from swyft.networks import OnlineNormalizationLayer
import torch.nn.functional as F
class HeadLowNoBAO(swyft.Module):
    def __init__(self, obs_shapes):
        super().__init__(obs_shapes=obs_shapes)

        self.n_features = 15
        self.l = torch.nn.Linear(7497, 10)
        self.l2 = torch.nn.Linear(98,5)
        self.onl_f = OnlineNormalizationLayer(torch.Size([7497]))
        self.onl_f2 = OnlineNormalizationLayer(torch.Size([98]))

    def forward(self, obs):
        x = torch.cat((obs['TT'][:,49:], obs['EE'][:,49:], obs['TE']), axis=1)
        x1 = torch.sqrt(torch.cat((obs['TT'][:,:49],
                        obs['EE'][:,:49]), axis=1))
        x=torch.cat((x,x1),axis=1)
        x1=torch.sqrt(x1)
        x=self.onl_f(x)
        x1=self.onl_f2(x1)
        nbatch = len(obs)
        x = self.l(x)
        x1=self.l2(x1)

        return torch.cat((x,x1),axis=1)
        #return x
```

```python
# default linear compression head.
from swyft.networks import OnlineNormalizationLayer
class HeadLinear(swyft.Module):
    def __init__(self, obs_shapes):
        super().__init__(obs_shapes=obs_shapes)

        self.n_features = 10
        self.l = torch.nn.Linear(7497, 10)
        self.onl_f = OnlineNormalizationLayer(torch.Size([7497]))

    def forward(self, obs):
        x = torch.cat((obs["TT"], obs['EE'],obs['TE']), axis=1)
        x=self.onl_f(x)
        x = self.l(x)
        
        return x
```

modified plotting routine so we can overlay samples.

```python
def corner(
    samples,
    params,
    fig=None,
    axes=None,
    figsize=(10, 10),
    color="k",
    labels=None,
    label_args={},
    truth=None,
    bins=100,
):
    K = len(params)
    if fig is None or axes is None:
        fig, axes = plt.subplots(K, K, figsize=figsize)
    lb = 0.125
    tr = 0.9
    whspace = 0.1
    fig.subplots_adjust(
        left=lb, bottom=lb, right=tr, top=tr, wspace=whspace, hspace=whspace
    )

    if labels is None:
        labels = [params[i] for i in range(K)]
    for i in range(K):
        for j in range(K):
            ax = axes[i, j]
            # Switch off upper left triangle
            if i < j:
                ax.set_yticklabels([])
                ax.set_xticklabels([])
                ax.set_xticks([])
                ax.set_yticks([])
                ax.set_frame_on(False)
                continue

            # Formatting labels
            if j > 0 or i == 0:
                ax.set_yticklabels([])
                # ax.set_yticks([])
            if i < K - 1:
                ax.set_xticklabels([])
                # ax.set_xticks([])
            if i == K - 1:
                ax.set_xlabel(labels[j], **label_args)
            if j == 0 and i > 0:
                ax.set_ylabel(labels[i], **label_args)

            # Set limits
            # ax.set_xlim(x_lims[j])
            # if i != j:
            #    ax.set_ylim(y_lims[i])

            # 2-dim plots
            if j < i:
                swyft.utils.plot_posterior(
                    samples, [params[j], params[i]], ax=ax, color=color, bins=bins, alpha=0
                )
                if truth is not None:
                    ax.axvline(truth[params[j]], color="r")
                    ax.axhline(truth[params[i]], color="r")
            if j == i:
                swyft.utils.plot_posterior(samples, params[i], ax=ax, color=color, bins=bins)
                if truth is not None:
                    ax.axvline(truth[params[i]], ls=":", color="r")
    return fig, axes
```

# empirical credibility test

for technical reasons, if loading task from file, need to use "noise" version of code, commented out currently. If you get very ugly results, it's likely the wrong line is being used.

```python
from tqdm import tqdm
def estimate_empirical_mass(dataset, post, nobs, npost):
    obs0, v0 = dataset[0]
    obs0 = {k: v.numpy() for k,v in obs0.items()}
    #obs0 = noise({k: v.numpy() for k,v in obs0.items()})
    w0 = {k: np.exp(v) for k,v in post._eval_ratios(obs0,v0.unsqueeze(0).numpy()).items()}
    mass = {
        k: dict(nominal=[], empirical=np.linspace(1 / nobs, 1, nobs)) for k in w0.keys()
    }
    for _ in tqdm(range(nobs)):
        j = np.random.randint(len(dataset))
        obs0, v0 = dataset[j]
        obs0 = {k: v.numpy() for k,v in obs0.items()} 
        #obs0 = noise({k: v.numpy() for k,v in obs0.items()})
        w0 =  {k: np.exp(v) for k,v in post._eval_ratios(obs0,v0.unsqueeze(0).numpy()).items()}
        wS = post.sample(npost, obs0)["weights"]
        for k, v in w0.items():
            f = wS[k][wS[k] >= v].sum() / wS[k].sum()
            mass[k]["nominal"].append(f)
    for k in mass.keys():
        mass[k]["nominal"] = np.asarray(sorted(mass[k]["nominal"]))

    return mass
```

```python
def empirical_mass(
        self, nobs: int = 1000, npost: int = 1000
    ):
        """Estimate empirical vs nominal mass.
        Args:
            nobs: Number of mock observations for empirical mass estimate (taken randomly from dataset)
            npost: Number of posterior samples to estimate nominal mass
        Returns:
            Nominal and empirical masses.
        """
        return estimate_empirical_mass(self.dataset, self.posteriors, nobs, npost)
```

```python
import scipy
```

```python
probit = lambda x: scipy.stats.norm.ppf(x)
z_from_alpha = lambda alpha: probit(1-alpha/2)
alpha_from_z = lambda z: 2-scipy.stats.norm.cdf(z)*2
alpha_from_z(2)
```

```python
def Jefferys_interval(k, n, z = 1):
    alpha = alpha_from_z(z=z)
    lower = scipy.stats.beta.ppf(alpha/2, k+0.5, n-k+0.5)
    upper = scipy.stats.beta.ppf(1-alpha/2, k+0.5, n-k+0.5)
    return np.array([np.where(k>0, lower, 0.), np.where(k < n, upper, 1.)]).T
```

```python
def estimate_hat_z(masses, nbins = 50, zmax = 4, z_band = 1):
    n = len(masses)
    zlist = np.linspace(0, zmax, nbins)
    tlist = 1-alpha_from_z(zlist)
    k = np.array([sum(masses < t) for t in tlist])
    r_mean = k/n    
    r_band = Jefferys_interval(k, n, z=z_band)
    z_mean = z_from_alpha(1-r_mean)
    z_band = z_from_alpha(1-r_band)
    return dict(z = zlist, mean = z_mean, upper = z_band[:,1], lower = z_band[:,0])
```

```python
def containmentPlot(pois,figsize=(10,10),fig=None, label_args={}):
    K = len(pois)
    if fig is None:
        fig, axes = plt.subplots(K, K, figsize=figsize)
    else:
        axes = np.array(fig.get_axes()).reshape((K, K))
    lb = 0.125
    tr = 0.9
    whspace = 0.1
    fig.subplots_adjust(
        left=lb, bottom=lb, right=tr, top=tr, wspace=whspace, hspace=whspace
    )

    diagnostics = {}

    for i in range(K):
        for j in range(K):
            ax = axes[i, j]
            # Switch off upper left triangle
            if i < j:
                ax.set_yticklabels([])
                ax.set_xticklabels([])
                ax.set_xticks([])
                ax.set_yticks([])
                ax.set_frame_on(False)
                continue

            # Formatting labels
            if j > 0 or i == 0:
                ax.set_yticklabels([])
                # ax.set_yticks([])
            if i < K - 1:
                ax.set_xticklabels([])
                # ax.set_xticks([])
            if i == K - 1:
                #ax.set_xlabel('$\\theta_'+str(j)+'$', **label_args)#labels[j], **label_args)
                ax.set_xlabel(Lnames[j], **label_args)#labels[j], **label_args)
            if j == 0: #and i > 0:
                ax.set_ylabel(Lnames[i], **label_args)

            # Set limits
            # ax.set_xlim(x_lims[j])
            # if i != j:
            #    ax.set_ylim(y_lims[i])

            # 2-dim plots
            if j < i:
                plot_band(masses[(j,i)]['nominal'],ax, zmax = 3.5)
            if j == i:
                plot_band(masses[(j,)]['nominal'],ax, zmax = 3.5)
        #fig.suptitle('Empirical Credibility vs. Nominal Credibility', fontsize=18)
    plt.subplots_adjust(wspace=0.1, hspace=0.1)
    return fig
```

# Neutrino mass

we consider the simplest case, where the three neutrinos have the same mass.

```python
lmax=2500
fsky=0.6
ell=np.array([l for l in range(2,lmax+1)])
ells=ell*(ell+1)/(2*np.pi)
#detector contribution to signal, from file
Nltt=fpr2[ell-2,1]
Nlee=fpr2[ell-2,2]

sigBAO=np.array([0.015,0.17,0.13,0.16])
bao0={'6DF': [0.106, 0.327,0.015, 7], 'LOWZ': [0.32,8.47, 0.17,3],
            'CMASS':   [0.57,13.77,0.13,3], 'MGS':[0.15,4.47,0.16,3]}
def simulatorNu(v,sigma=0.1):
    #to avoid some CosmoComputationErrors
    V=np.around(v, decimals=6)
    omega_b, omega_cdm, theta, lnAs,n_s,tau_reio, Mnu =V[0],V[1],V[2],V[3],V[4],V[5], V[6]
    Mnustr=str(Mnu/3)+', '+str(Mnu/3)+', '+str(Mnu/3)
    A_s=np.exp(lnAs)/ 1.e10
    params = {
        'output': 'tCl pCl lCl',
        'l_max_scalars': lmax,
        'lensing': 'yes',
        'omega_b': omega_b,
        'omega_cdm': omega_cdm,
        '100*theta_s': theta,
        'A_s': A_s,
        'n_s': n_s,
        'tau_reio': tau_reio,
        'N_ur': 0.00441,
        'N_ncdm': 1,
        'm_ncdm': Mnu/3,
        'deg_ncdm': 3
    }

    # Create an instance of the CLASS wrapper
    cosmo = Class()

    # Set the parameters to the cosmological code
    cosmo.set(params)

    # Run the code.
    try:
        cosmo.compute(['lensing'])
        # Extract observables.
        cls = cosmo.lensed_cl(lmax)
        T=cosmo.T_cmb()
        baoT=[]
        for i,k in enumerate(bao0.keys()):
            z=bao0[k][0]
            da = cosmo.angular_distance(z)
            dr = z / cosmo.Hubble(z)
            dv = pow(da * da * (1 + z) * (1 + z) * dr, 1. / 3.)
            rs = cosmo.rs_drag()

            if bao0[k][-1] == 3:
                theo = dv / rs
            elif bao0[k][-1] == 4:
                theo = dv
            elif bao0[k][-1] == 5:
                theo = da / rs
            elif bao0[k][-1] == 6:
                theo = 1. / cosmo.Hubble(z) / rs
            elif bao0[k][-1] == 7:
                theo = rs / dv
            baoT.append(theo)
        cosmo.struct_cleanup()
        return dict(TT=cls['tt'][ell]*(T*1.e6)**2+Nltt, TE=cls['te'][ell]*(T*1.e6)**2,\
                EE=cls['ee'][ell]*(T*1.e6)**2+Nlee, BAO=np.array(baoT))

    except:
            print("a simulation failed! Find a 3 in store.get_simulation_status")
            return dict(TT=np.nan, TE=np.nan, EE=np.nan, BAO=np.nan)
```
